```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random

#open the file
datapath='/Users/rachellee/Desktop/research/references/'
filename='flux_table_70um.txt'
data=np.loadtxt(datapath+filename)

#number is coloumn 0, flux is column 25
number=data[:,0]
flux=data[:,25]
high=np.max(flux)
index_max=np.argmax(flux)
print('Max. flux:',high,'at index',index_max)

low=np.min(flux)
index_min=np.argmin(flux)
print('Min. flux:',low,'at index',index_min)

fig,ax=plt.subplots()
ax.hist(flux,color='black')
ax.set_ylabel('Count',font='serif',fontsize=12)
ax.set_xlabel(r'70 $\mu$m Flux',font='serif',fontsize=12)
ax.set_title('Flux Distribution of Protostars (Furlan et al. 2016)',font='serif',fontsize=15)
plt.plot()
```

```python
flux_new=[]
for i in range(len(flux)):
    if flux[i]!=0:
        flux_new.append(flux[i])
```

```python
#assign a star a random value from the fluxes 
cluster_flux=[]
star=[]
for i in range(1000):
    cluster_flux.append(random.choice(flux_new))
    star.append(i)
    
fig,ax=plt.subplots()
ax.hist(cluster_flux,color='black')
ax.set_ylabel('Count',font='serif',fontsize=12)
ax.set_xlabel(r'70 $\mu$m Flux',font='serif',fontsize=12)
ax.set_title('Flux Distribution for Protostar Cluster (1000 stars)',font='serif',fontsize=15)
plt.plot()

fig,ax=plt.subplots()
ax.plot(star,cluster_flux,color='black')
ax.set_ylabel(r'70 $\mu$m Flux',font='serif',fontsize=12)
ax.set_xlabel('Star Number',font='serif',fontsize=12)
ax.set_title('Flux Distribution for Protostar Cluster (1000 stars)',font='serif',fontsize=15)
plt.plot()
```

```python
#convert flux to mag
cluster_mag=[]
for i in range(len(cluster_flux)):
    mag=-2.5*np.log10(cluster_flux[i])
    cluster_mag.append(mag)
```

```python
#plot in mag
fig,ax=plt.subplots()
ax.scatter(star,cluster_mag,color='black')
ax.set_ylabel('Magnitude',font='serif',fontsize=12)
ax.set_xlabel('Protostar Number',font='serif',fontsize=12)
ax.set_title('Magnitude Distribution for Protostar Cluster (1000 stars)',font='serif',fontsize=15)
ax.invert_yaxis()
plt.plot()

fig,ax=plt.subplots()
ax.hist(cluster_mag,color='black')
ax.set_ylabel('Count',font='serif',fontsize=12)
ax.set_xlabel(r'70 $\mu$m Flux',font='serif',fontsize=12)
ax.set_title('Flux Distribution for Protostar Cluster (1000 stars)',font='serif',fontsize=15)
plt.plot()
```

```python
#choose a random star in the set to burst
cluster_new=[]
for i in range(len(cluster_mag)):
    choice_index=random.randrange(len(cluster_mag))
    new_mag=cluster_mag[choice_index]-5
    cluster_mag[choice_index]=new_mag
    cluster_new=cluster_mag
    
fig,ax=plt.subplots()
ax.plot(star,cluster_new,color='black')
ax.set_ylabel('Magnitude',font='serif',fontsize=12)
ax.set_xlabel('Protostar Number',font='serif',fontsize=12)
ax.set_title('Magnitude Distribution for Protostar Cluster (1000 stars)',font='serif',fontsize=15)
ax.invert_yaxis()
plt.plot()

fig,ax=plt.subplots()
ax.hist(cluster_new,color='black')
ax.set_ylabel('Count',font='serif',fontsize=12)
ax.set_xlabel('Magnitude',font='serif',fontsize=12)
ax.set_title('Magnitude Distribution for Protostar Cluster (1000 stars)',font='serif',fontsize=15)
plt.plot()
```

```python
cluster_new=[]
star_new=[]
cluster_burst=[]
star_burst=[]
choice_index=random.randrange(len(cluster_mag))
print(choice_index)
for i in range(len(cluster_mag)+1):
    if i==choice_index:
        new_mag=cluster_mag[i]-5
        cluster_burst.append(new_mag)
        star_burst.append(star[i])
        cluster_new=cluster_mag.pop(i)
        star_new=star.pop(i)

fig,ax=plt.subplots(1,1)
fig.set_size_inches(7,6)
ax.scatter(star,cluster_mag,color='black',alpha=0.3,s=25)
ax.scatter(star_new,cluster_new,color='blue',label='after accretion event',alpha=1,s=50)
ax.scatter(star_burst,cluster_burst,color='red',label='before accretion event',alpha=1,s=50)
ax.set_xlim(-10,1010)
ax.set_ylim(-9,8)
ax.set_title('Comparing the cluster before and after an accretion event',font='serif',fontsize=15)
ax.set_xlabel('Star number',font='serif',fontsize=12)
ax.set_ylabel('Absolute magnitude',font='serif',fontsize=12)
ax.legend(loc='best')
plt.gca().invert_yaxis()
```
