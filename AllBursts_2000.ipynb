```python
run=50
t=300
cadence=26

#run it 50 times
#start with function
def bursts1(full_time,size,probability,L_matrix,burst_matrix,change_matrix,new_burst):
    cluster_flux=[]
    star=[]
    cluster_size=size
    for i in range(cluster_size):
        cluster_flux.append(random.choice(flux_new))
        star.append(i)
    cluster_mag=[]
    for i in range(len(cluster_flux)):
        mag=-2.5*np.log10(cluster_flux[i])
        cluster_mag.append(mag)
    #make multi dimensional array for burst prob 
    #let 1 be unburst and NaN be burst
    t=full_time #yrs
    cadence=26
    steps=t*cadence
    burst_prob=probability/cadence
    burst=burst_matrix
    L=L_matrix
    C=change_matrix
    N=new_burst
    for i in range(steps):
        for j in range(cluster_size):
            if burst[j,i]==1:
                prob1=random.random()
                if prob1>burst_prob:
                    burst[j,i]=burst[j,i]
                    L[j,i]=cluster_mag[j]
                    C[j,i]=0
                    N[j,i]=0
                elif prob1<=burst_prob:
                    burst[j,i:(30*cadence)+i]=np.nan
                    L[j,i:(30*cadence)+i]=cluster_mag[j]-5
                    C[j,i:(30*cadence)+i]=100
                    N[j,i]=100
                    N[j,i+(30*cadence)]=100
    #determine total number of bursts 
    total=0
    burst_total=[]
    for i in range(steps):
        for j in range(cluster_size):
            if np.isnan(burst[j,i])==True and burst[j,i-1]==1:
                    total+=1
        burst_total.append(total)
    #find the fraction at a time
    burst_num=[]
    fraction_val=0
    for i in range(steps):
        for j in range(cluster_size):
            if np.isnan(burst[j,i])==True:
                fraction_val+=1
        burst_num.append(fraction_val)
        fraction_val=0
        
    return burst_total,burst_num,burst,L,C,new_burst
```

```python
cluster_size2000_1=2000
steps=t*cadence
burst2000_1=np.ones((cluster_size2000_1,8580))
L2000_1=np.zeros((cluster_size2000_1,8580))
C2000_1=np.empty((cluster_size2000_1,8580))
new_burst_1=np.zeros((cluster_size2000_1,8580))
prob2000_1=1/(101*30)

#1 bin
burst_total2000_1=[]
burst_num2000_1=[]
final_burst2000_1=[]
final_L2000_1=[]
final_C2000_1=[]
final_2000_1=[]

for i in range(run):
    print(i)
    burst2000_1=np.ones((cluster_size2000_1,8580))
    L2000_1=np.zeros((cluster_size2000_1,8580))
    C2000_1=np.empty((cluster_size2000_1,8580))
    new_burst_1=np.zeros((cluster_size2000_1,8580))
    values2000_1=bursts1(t,2000,prob2000_1,L2000_1,burst2000_1,C2000_1,new_burst_1)
    burst_total2000_1.append(values2000_1[0])
    burst_num2000_1.append(values2000_1[1])
    final_burst2000_1.append(values2000_1[2])
    final_L2000_1.append(values2000_1[3])
    final_C2000_1.append(values2000_1[4])
    final_2000_1.append(values2000_1[5])
```

```python
final_burst2000_1bin=np.empty((run,cluster_size2000_1,steps))
final_L2000_1bin=np.empty((run,cluster_size2000_1,steps))
final_C2000_1bin=np.empty((run,cluster_size2000_1,steps))
final_2000_1bin=np.empty((run,cluster_size2000_1,steps))
for i in range(run):
    for j in range(cluster_size2000_1):
        for k in range(steps):
            final_burst2000_1bin[i,j,k]=final_burst2000_1[i][j][k]
            final_L2000_1bin[i,j,k]=final_L2000_1[i][j][k]
            final_C2000_1bin[i,j,k]=final_C2000_1[i][j][k]
            final_2000_1bin[i,j,k]=final_2000_1[i][j][k]
```
