```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random

#open the file
datapath='/Users/rachellee/Desktop/research/references/'
filename='flux_table_70um.txt'
data=np.loadtxt(datapath+filename)

#number is coloumn 0, flux is column 25
number=data[:,0]
flux=data[:,25]
```

```python
#remove zeros in the flux
flux_new=[]
for i in range(len(flux)):
    if flux[i]!=0:
        flux_new.append(flux[i])
```

# 1000 stars

```python
#assign a star a random value from the fluxes 
cluster_flux1000=[]
star1000=[]
cluster_size1000=1000
for i in range(cluster_size1000):
    cluster_flux1000.append(random.choice(flux_new))
    star1000.append(i)
```

```python
#convert flux to mag
cluster_mag1000=[]
for i in range(len(cluster_flux1000)):
    mag1000=-2.5*np.log10(cluster_flux1000[i])
    cluster_mag1000.append(mag1000)
```

```python
#make multi dimensional array for burst prob 
#let 1 be unburst and NaN be burst
t=250 #yrs
cadence=2
steps=t*cadence
burst_prob=1e-4/cadence
burst1000=np.ones((cluster_size1000,steps))
L1000=np.zeros((cluster_size1000,steps))
```

```python
#set up a loop that sets the star that bursts to NaN for 100 years
burst_fraction1000=[]
cluster_new1000=[]
cluster_old1000=[]
star_new1000=[]
star_old1000=[]
bursts1000=0
for i in range(steps):
    for j in range(cluster_size1000):
        if burst1000[j,i]==1:
            prob=random.random()
            if prob>burst_prob:
                burst1000[j,i]=burst1000[j,i]
                L1000[j,i]=cluster_mag1000[j]
            if prob<=burst_prob:
                burst1000[j,i:100*cadence+i]=np.nan
                L1000[j,i]=cluster_mag1000[j]-5
                cluster_new1000.append(cluster_mag1000[j]-5)
                cluster_old1000.append(cluster_mag1000[j])
                star_new1000.append(star1000[j])
                star_old1000.append(star1000[j])
```

```python
#determine total number of bursts 
total1000=0
burst_total1000=[]
for i in range(steps):
    for j in range(cluster_size1000):
        if np.isnan(burst1000[j,i])==True and burst1000[j,i-1]==1:
                total1000+=1
    burst_total1000.append(total1000)
```

```python
#find the fraction at a time
burst_fraction1000=[]
fraction_val1000=0
for i in range(steps):
    for j in range(cluster_size1000):
        if np.isnan(burst1000[j,i])==True:
            fraction_val1000+=1
    burst_fraction1000.append(fraction_val1000)
    fraction_val1000=0
```

```python
fraction1000=[]
for i in range(len(burst_fraction1000)):
    fraction1000.append(burst_fraction1000[i]/cluster_size1000)
```

```python
time=np.linspace(0,t,steps)
text="t={} years".format(t)
```

```python
title4='Total Number Bursts in Cluster of Size {}'.format(cluster_size1000)
title5='Fraction of Stars Bursting at a Time in Cluster of Size {}'.format(cluster_size1000)
fig,(ax1,ax2)=plt.subplots(1,2)
fig.set_size_inches(16,7)
ax1.plot(time,burst_total1000,color='black')
ax1.set_xlabel('Time [years]',font='serif',fontsize=13)
ax1.set_ylabel('Bursts [Number of Protostars]',font='serif',fontsize=13)
ax1.set_title(title4,font='serif',fontsize=15)
ax1.text(t,0,text,font='serif',fontsize=12,ha='right')
ax2.plot(time,fraction1000,color='black')
ax2.set_xlabel('Time [years]',font='serif',fontsize=13)
ax2.set_ylabel('Fraction',font='serif',fontsize=13)
ax2.set_title(title5,font='serif',fontsize=15)
ax2.text(t,0.0005,text,font='serif',fontsize=12,ha='right')
#plt.savefig('/Users/rachellee/Desktop/comp_bursts_normal1000.jpeg',dpi=300)
```

# mean and standard deviation for different cluster sizes

```python
#first observation at t=200 years
#PRIMA starts 20 years later and observes once every 6 months for 5 years
t_list_years=[200,220,220.5,221,221.5,222,222.5,223,223.5,224,224.5,225]
t_list_cadence=[int(i*2) for i in t_list_years]
t_fraction1000=[]
t_fraction50=[]
t_fraction100=[]
t_fraction500=[]
t_fraction2000=[]
t_fraction5000=[]
for i in t_list_cadence:
    t_fraction50.append(fraction50[i])
    t_fraction100.append(fraction100[i])
    t_fraction500.append(fraction500[i])
    t_fraction1000.append(fraction1000[i])
    t_fraction2000.append(fraction2000[i])
    t_fraction5000.append(fraction5000[i])
```

```python
mean50=np.mean(t_fraction50)
std50=np.std(t_fraction50)
mean100=np.mean(t_fraction100)
std100=np.std(t_fraction100)
mean500=np.mean(t_fraction500)
std500=np.std(t_fraction500)
mean1000=np.mean(t_fraction1000)
std1000=np.std(t_fraction1000)
mean2000=np.mean(t_fraction2000)
std2000=np.std(t_fraction2000)
mean5000=np.mean(t_fraction5000)
std5000=np.std(t_fraction5000)

mean=[mean50,mean100,mean500,mean1000,mean2000,mean5000]
std=[std50,std100,std500,std1000,std2000,std5000]
number=[50,100,500,1000,2000,5000]
```

```python
fig,ax=plt.subplots()
fig.set_size_inches(8,7)
ax.errorbar(number,mean,yerr=std,fmt='|',color='black',ecolor='black',elinewidth=2)
ax.text(5000,-0.004,text,font='serif',fontsize=12,ha='right')
ax.set_xlabel('Number of Stars in Cluster',font='serif',fontsize=12)
ax.set_ylabel('Fraction of Bursts',font='serif',fontsize=12)
ax.axhline(0.01,color='red')
ax.set_title('Fraction of Stars with FU Ori Event for each Observation',font='serif',fontsize=15)
#plt.savefig('/Users/rachellee/Desktop/10Run_SizeChange.jpeg',dpi=300)
```

```python
from tabulate import tabulate
table=[['Cluster Size','Mean fraction of bursts','Standard deviation'],
      [50,np.round(mean50,4),np.round(std50,4)],
      [100,np.round(mean100,4),np.round(std100,4)],
      [500,np.round(mean500,4),np.round(std500,4)],
      [1000,np.round(mean1000,4),np.round(std1000,4)],
       [2000,np.round(mean2000,4),np.round(std2000,4)],
      [5000,np.round(mean5000,4),np.round(std5000,4)]]
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid',numalign='center'))
```
