```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.optimize import curve_fit
from matplotlib.patches import Ellipse
from matplotlib.patches import Rectangle
```

```python
duration=[10**4,10**3,10**3,10**(2.5),10**(0.95),10**0]
mag=[5,4,2,3,1.25,0.5]

plt.scatter(duration,mag)
plt.xscale('log')
```

```python
def power_law(x,a,b):
    return a*(x**b)

pars_mag,cov_mag=curve_fit(f=power_law,xdata=duration,ydata=mag,p0=[0,0],bounds=(-np.inf,np.inf))
stdevs_mag=np.sqrt(np.diag(cov_mag))
res_mag=mag-power_law(duration,*pars_mag)
```

```python
plt.scatter(duration,mag)
plt.xscale('log')
plt.plot(duration,power_law(duration,*pars_mag),linestyle='--',linewidth=1.5,color='black')
```

```python
print('a=',pars_mag[0])
print('b=',pars_mag[1])
```

```python
flux=[]
for i in range(len(mag)):
    f=100**(mag[i]/5)
    flux.append(f)
print(flux)
```

```python
pars_flux,cov_flux=curve_fit(f=power_law,xdata=duration,ydata=flux,p0=[0,0],bounds=(-np.inf,np.inf))
res_flux=flux-power_law(duration,*pars_flux)
```

```python
plt.scatter(duration,flux)
plt.xscale('log')
plt.plot(duration, power_law(duration, *pars_flux), linestyle='--', linewidth=1.5, color='black')
```

```python
print('a=',pars_flux[0])
print('b=',pars_flux[1])
```

```python
dur_days1=(5/pars_mag[0])**(1/pars_mag[1])
dur_years1=dur_days1/365
print(dur_days1)

dur_days2=(100/pars_flux[0])**(1/pars_flux[1])
dur_years2=dur_days2/365
print(dur_days2)
```

```python
mag=pars_mag[0]*((100*365)**pars_mag[1])
print(mag)
```
